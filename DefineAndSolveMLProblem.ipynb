```python
y = df['Perceptions of corruption']
X = df.drop(columns = 'Perceptions of corruption', axis = 1)
```

```python
# creating the testing and training data sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=1234)
```

```python
# Decision Tree Regressor
md = [4, 8]
msl = [25, 50]
param_grid={'max_depth':md, 'min_samples_leaf':msl}
```

```python
# running a grid search CV
print("Running Grid Search CV...")

dt_regressor = DecisionTreeRegressor()

dt_grid = GridSearchCV(dt_regressor, param_grid, cv=3,scoring='neg_root_mean_squared_error', verbose=1)

dt_grid_search = dt_grid.fit(X_train, y_train)
print('Done')
```

```python
# finding the RSME vale of the best model
rmse_DT = -1 * dt_grid_search.best_score_
print("[DT] RMSE for the best model is : {:.2f}".format(rmse_DT) )
```

```python
dt_best_params = dt_grid_search.best_params_
dt_best_params
```

```python
# fitting the model on the best params
dt_model = DecisionTreeRegressor(max_depth = dt_best_params['max_depth'],
min_samples_leaf = dt_best_params['min_samples_leaf'])
dt_model.fit(X_train, y_train)
```

```python
# Testing it on testing set to get a prediction

# prediction
y_dt_pred = dt_model.predict(X_test)

# computing the RSME
dt_rmse = mean_squared_error(y_test, y_dt_pred, squared=False)

# computing r2
dt_r2 = r2_score(y_test, y_dt_pred)

print('[DT] Root Mean Squared Error: {0}'.format(dt_rmse))
print('[DT] R2: {0}'.format(dt_r2))
```

```python
X.head()
```

```python
# getting the features of most importance
feature_importance = dt_model.feature_importances_
feature_importance_df = pd.DataFrame({'Feature' : X.columns, 'Importance': feature_importance})
feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)

# creating a barplot using matplotlib
plt.figure(figsize=(12, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title("Decision Tree Regressor Feature Importance")
plt.show()

```

```python
# removing features with little to no importance
df = df.drop(columns = ['Life Ladder', 'Standard deviation/Mean of ladder by country-year', 'Healthy life expectancy at birth', 'Freedom to make life choices', 'Generosity', 'gini of household income reported in Gallup, by wp5-year'], axis = 0)
```

```python
features = df.loc[:, df.columns != 'Perceptins of corruption']
label = df['Perceptions of corruption']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=1234)
```

```python
dt_model = DecisionTreeRegressor(max_depth = dt_best_params['max_depth'],
min_samples_leaf = dt_best_params['min_samples_leaf'])
dt_model.fit(X_train, y_train)
```

```python
# Testing it on testing set to get a prediction

# prediction
y_dt_pred = dt_model.predict(X_test)

# computing the RSME
dt_rmse = mean_squared_error(y_test, y_dt_pred, squared=False)

# computing r2
dt_r2 = r2_score(y_test, y_dt_pred)

print('[DT] Root Mean Squared Error: {0}'.format(dt_rmse))
print('[DT] R2: {0}'.format(dt_r2))
```

```python
# Trying a randomforest model to see if it'll make the R2 and RSME better
rf_model = RandomForestRegressor()

rf_model.fit(X_train, y_train)

prediction = rf_model.predict(X_test)
prediction
```

```python
rf_rmse = mean_squared_error(y_test, prediction, squared=False)

# computing r2
rf_r2 = r2_score(y_test, prediction)

print('[DT] Root Mean Squared Error: {0}'.format(rf_rmse))
print('[DT] R2: {0}'.format(rf_r2))
```

```python
# This is a lot better than the one from the DT Regressor
# Optimizing it using grid search CV

pipeline = Pipeline([('model', RandomForestRegressor(random_state=0))])

param_grid = {
    'model__n_estimators' : [100, 200, 300],
    'model__max_depth' : [None, 10, 20, 30],
    'model__min_samples_split' : [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4],
    'model__max_features': ['auto', 'sqrt', 'log2']}

grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv =3, scoring='neg_mean_squared_error', n_jobs=1, verbose=2)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best parameters found ", best_params)
```

```python
# new model with best_params
rf2_model = RandomForestRegressor(max_depth=30, max_features = 'sqrt', min_samples_leaf = 1, min_samples_split = 2, n_estimators=200)

rf2_model.fit(X_train, y_train)

prediction2 = rf2_model.predict(X_test)
prediction2
```

```python
rf2_rmse = mean_squared_error(y_test, prediction2, squared=False)

# computing r2
rf2_r2 = r2_score(y_test, prediction2)

print('[DT] Root Mean Squared Error: {0}'.format(rf2_rmse))
print('[DT] R2: {0}'.format(rf2_r2))
```
